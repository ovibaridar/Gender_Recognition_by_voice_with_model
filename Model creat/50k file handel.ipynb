```python
import shutil
import os



def copy_and_rename_files(df, source_folder, destination_folder, csv_filename):
    renamed_files = []  # List to store renamed file names
    
    for index, row in df.iterrows():
        source_file = os.path.join(source_folder, row['path'])
        new_file_name = str(index + 1) + os.path.splitext(row['path'])[1]
        destination_file = os.path.join(destination_folder, new_file_name)
        
        try:
            shutil.copy(source_file, destination_file)
            print(f"File '{source_file}' copied and renamed to '{destination_file}' successfully.")
            renamed_files.append({'original_filename': row['path'], 'renamed_filename': new_file_name})
        except IOError as e:
            print(f"Unable to copy file '{source_file}'. Error: {e}")

    # Save renamed file names to CSV
    renamed_df = pd.DataFrame(renamed_files)
    renamed_df.to_csv(csv_filename, index=False)
    print(f"Renamed file names saved to '{csv_filename}'.")

# Example usage:
source_folder = r"H:\data\cv-corpus-16.1-delta-2023-12-06\en\clips"
destination_folder = r"H:\data\cv-corpus-16.1-delta-2023-12-06\en\Male"
csv_filename = "renamed_files2.csv"

copy_and_rename_files(male, source_folder, destination_folder, csv_filename)
```
